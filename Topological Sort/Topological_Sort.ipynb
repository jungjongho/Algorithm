```python
class Node(object):

    def __init__(self,x):
        self.data=x
        self.inlayer=[]
        self.outlayer=[]

    def insert_out(self,x):
        self.outlayer.append(x)

    def insert_in(self,x):
        self.inlayer.append(x)

class Graph(object):
    def __init__(self):
        self.node=[]

    def insert_node(self,x):
        self.node.append(x)

    def topological_sort(self):
        list_graph=self.node
        queue=[]
        sort=[]
        for _ in range(len(list_graph)):
            for i in list_graph:
                if len(i.inlayer)==0:
                    queue.append(i)
                    sort.append(i.data)
                    self.node.remove(i)

                for j in list_graph:
                    for i in queue:
                        if i.data in j.inlayer:
                            j.inlayer.remove(i.data)

                    
        print(sort)
```
